```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings(action="ignore")
```

```python
data = {
    "person":["Person 1","Person 2","Person 3","Person 4","Person 5","Person 6"],
    "QSA1":[20,135,75,23,70,140],
    "QSA2":[25,150,50,20,42,135]
}

df = pd.DataFrame(data)
print(df)
```

```python
plt.scatter(df["QSA1"], df["QSA2"])
plt.xlabel("QSA1")
plt.ylabel("QSA2")
plt.title("QUARTER SPENDDING AMOUNT")

plt.show()
```

```python
df
```

```python
df = df[["QSA1","QSA2"]]
df
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(df)
```

```python
df["Cluster"] = kmeans.labels_
print(df)
```

```python
print("Centroids:")

print(kmeans.cluster_centers_)
```

```python
plt.scatter(df["QSA1"], df["QSA2"],
            c=df["Cluster"],
            cmap="viridis",
            s=100)

plt.scatter(kmeans.cluster_centers_[:,0],
            kmeans.cluster_centers_[:,1],
            color="red",
            marker="X",
            s=250)

plt.xlabel("QSA1")
plt.ylabel("QSA2")
plt.title("K-Means Clustering")
plt.show()
```
